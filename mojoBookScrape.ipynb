```python
from selenium import webdriver
```

```python
driver=webdriver.Chrome()
```

```python
driver.get('https://www.boxofficemojo.com/chart/top_lifetime_gross/?area=XWW')
```

```python
indexes = driver.find_elements("xpath","//td[@class='a-text-right mojo-header-column mojo-truncate mojo-field-type-rank']")
```

```python
for index in indexes:
    print(index.text)
```

```python
titles = driver.find_elements("xpath", "//td[@class='a-text-left mojo-field-type-title']")
```

```python
for title in titles:
    print(title.text)
```

```python
prices = driver.find_elements("xpath","//td[@class='a-text-right mojo-field-type-money']")
```

```python
for price in prices:
    print(price.text)
```

```python
years = driver.find_elements("xpath","//td[@class='a-text-left mojo-field-type-year']")
```

```python
for year in years:
    print(year.text)
```

```python
books =zip(titles, years, prices)
```

```python
import csv
with open ("D:\data preparing\datasets\books.csv" , "w" ,newline="" ,encoding="utf-8") as file :
    writer = csv.writer(file)
    writer.writerow(["Title", "Year", "Price"])
    writer.writerows(books)
```

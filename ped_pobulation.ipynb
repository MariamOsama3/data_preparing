**to make number with zeros not M or K**

```python

def value_to_float(x):
    if isinstance(x, (int, float)):  # Check if x is a number (int or float)
        return x
    if isinstance(x, str):  # Process only if x is a string
        if 'k' in x:
            return float(x.replace('k', '')) * 1000
        if 'M' in x:
            return float(x.replace('M', '')) * 1000000
        if 'B' in x:
            return float(x.replace('B', '')) * 1000000000

data.iloc[:, 1:] = data.iloc[:, 1:].applymap(value_to_float)

print(data)
```

**convert the number to be as float type**

```python
data.iloc[:,1:] = data.iloc[:,1:].astype('float')
```
